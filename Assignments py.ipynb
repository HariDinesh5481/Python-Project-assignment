```python
# Check if 97 is a prime number
num = 97
if num % 2 == 0:
    print(f"{num} is not a prime number")
elif num % 3 == 0:
    print(f"{num} is not a prime number")
elif num % 5 == 0:
    print(f"{num} is not a prime number")
elif num % 7 == 0:
    print(f"{num} is not a prime number")   
elif num % 11 == 0:
    print(f"{num} is not a prime number")   
else:
    print(f"{num} is a prime number")
```

```python
# Check if 97 is a prime number
num = 97
if num <= 1:
    print("It is a not a prime number")
else :
    is_prime = True
    for i in range(2, int(num**0.5)+1):
        if num % i ==0:
            is_prime = False
            break
if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# Find the factorial of 8 FOR LOOP
num = 8
factorial = 1
for i in range(1,num+1):
    factorial = factorial * i
print(f"The factorial of {num} is {factorial}")



```

```python
# Find the factorial of 8 WHILE LOOP
num = 8 
fact = 1
i=1
while i <= num:
    fact = fact * i
    i = i +1 
print(f"The factorial of {num} is {fact}")
```
