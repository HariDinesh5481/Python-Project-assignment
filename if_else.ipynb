```python
username = "admin"
password = "1234"

attempts = 0
max_attempts = 3
locked = False

while attempts < max_attempts:
    user = input("Username: ")
    pwd = input("Password: ")

    if locked:
        print("Account locked.")
        break

    if user == username and pwd == password:
        print("Login successful!")
        break
    else:
        attempts += 1
        remaining = max_attempts - attempts
        print("Wrong credentials.")

        if attempts == max_attempts:
            locked = True
            print("Account locked due to too many failed attempts.")
        else:
            print(f"{remaining} attempts left.")
```

```python
weight = float(input("Enter package weight (kg): "))
destination = input("Destination (domestic/international): ").lower()
shipping_type = input("Shipping type (standard/express): ").lower()

if destination == "domestic":
    rate = 5
elif destination == "international":
    rate = 15
else:
    print("Invalid destination")
    exit()

if shipping_type == "standard":
    multiplier = 1.0
elif shipping_type == "express":
    multiplier = 1.5
else:
    print("Invalid shipping type")
    exit()

cost = weight * rate * multiplier
print(f"Total Shipping Cost: ${cost:}")
```
